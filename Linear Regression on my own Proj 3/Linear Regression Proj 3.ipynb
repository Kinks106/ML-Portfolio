```python
#import llibraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Salary_dataset.csv')
```

```python
plt.scatter(df.YearsExperience,df.Salary)
plt.show()
```

```python
#loss function
```

```python
def loss_function(m,b,points):
    total_error = 0
    for i in range(len(points)):
        x=points.iloc[i].YearsExperience
        y=points.iloc[i].Salary
        total_error += (y - (m * x + b))**2
    total_error/float(len(points))
```

```python
#Gradient descent function
```

```python
def gradient_descent(m_now,b_now,points,learning_rate):
    m_gradient=0
    b_gradient=0

    n=len(points)

    for i in range(n):
        x=points.iloc[i].YearsExperience
        y=points.iloc[i].Salary

        m_gradient += -(2/n)*x*(y-(m_now*x+b_now))
        b_gradient += -(2/n)*(y-(m_now*x+b_now))

    m = m_now- m_gradient * learning_rate
    b = b_now- b_gradient * learning_rate
    return m,b
```

```python
m=0
b=0
L=0.0001
epochs=1000

for i in range(epochs):
    if i%50==0:
        print(f"epoch:{i}")
    m,b=gradient_descent(m,b,df,L)

print(m,b)
plt.scatter(df.YearsExperience,df.Salary,color="black")
plt.plot(list(range(2,12)),[m*x+b for x in range(2,12)],color="red")
plt.show()
```
